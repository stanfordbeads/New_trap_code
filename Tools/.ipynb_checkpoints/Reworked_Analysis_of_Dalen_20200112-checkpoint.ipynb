```python
def plot_basic_harmonic_sum(df_list,axis,norm_factor=1,no_harmonics=15,label="label",var_of_interest="attractor_position_z",output_file="test.png",save_file=True):
    "returns the harmonic sum for the data frame normed by the length of the data file"
    sum_list =[]
    for i in range(len(df_list)):
        _,sumws = get_mean_and_sum_of_harmonics(df_list[i],axis=axis,no_harmonics=no_harmonics,norm_factor=norm_factor)
        sum_list.append(sumws)
        plt.plot(np.divide(sumws,len(df_list[i])),linestyle="-",marker="o", label="%s: " %label + str(np.mean(df_list[i]["%s" %var_of_interest]))[:5])#np.mean(df_sub_list[i]["attractor_position_z"]))
        plt.yscale("log")
        plt.ylabel(r" sum NSD [AU/$\sqrt{Hz}$]")
        plt.xlabel("harmonic")
    plt.legend()
    if(save_file==True): plt.savefig(output_file, dpi=300, bbox_inches="tight")    
    plt.show()
```

```python
plot_basic_harmonic_sum(df_list[0:2],axis="x",norm_factor=1,no_harmonics=15,label="z",save_file=False)
```
