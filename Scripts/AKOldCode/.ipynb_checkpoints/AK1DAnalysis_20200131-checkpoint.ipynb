```python
#template+noise generator
for i in range(0,1):
    L=5.01e-6#pow(10,i/30.0-0.3)*1e-6
    D=8e-6
    H=6e-6
    Fshake=3
    test = force_vs_time(D,H,100e-6,Fshake,"z",L,yuk_or_grav="yuk",alpha=1e10)
    ofname = "/home/analysis_user/Results/AkioK/2020_01_31_BootsTrapping/SignalZD%dumH%dumL%.02f.txt" % (D*1e6,H*1e6,L*1e6)
    
    fsamp=5000
    res=50000
    Nint=2
    Nave=3
    norm = np.sqrt(2 / (res* fsamp))

    s = np.arange(0, 5, 0.0001)
    for i in range(0,10):
        for j in range(0,5000):
            s[j+i*5000]=test[1][j]*1e15+np.random.normal(0, 0.1)
    
    s_det=signal.detrend(s)
    
    sFFT=np.fft.rfft(s_det)
    spsd = (sFFT * sFFT.conj()).real
    spsd = norm**2 * (sFFT * sFFT.conj()).real
    snsd=np.sqrt(spsd)
    
    f1 = open(ofname,"a")
    
    for i in range(1,21):
        Ints=sFFT[Fshake*10*i]
        IntNSDs=snsd[Fshake*10*i]
    
        for j in range(1,(Nint)):
            Ints+=sFFT[Fshake*10*i+j]+sFFT[Fshake*10*i-j]
            IntNSDs+=snsd[Fshake*10*i+j]+snsd[Fshake*10*i-j]
 
        AveNoiseX=0
        SigAmpX=IntNSDs-AveNoiseX*(Nint*2+1)
        SigPhaX=np.angle(Ints)         
        f1.write('%f %e %e %e %e %e %e\n' % (Fshake*i, SigAmpX, 0,0, SigPhaX, 0,0))
    f1.close()
```
