```python
#for z force, 3/20 data: 6 um height, 9.05 um offset
for i in range(0,70):
    L=pow(10,i/30.0-0.3)*1e-6
    D=17e-6
    H=-6e-6
    Fshake=3
    test = force_vs_time(D,H,200e-6,Fshake,"z",L,yuk_or_grav="yuk",alpha=1e10,offset_y=-2.94e-6,bead_size=3.8e-6)
    ofname = "/data/new_trap/Results/AkioK/2020_05_18_Bead2/SignalZD%dumH%dumL%.02f.txt" % (D*1e6,H*1e6,L*1e6)
    
    
    fsamp=5000
    res=50000

    norm = np.sqrt(2 / (res* fsamp))

    s = np.arange(0, 5, 0.0001)
    for i in range(0,10):
        for j in range(0,5000):
            s[j+i*5000]=test[1][j]
    
    sFFT=np.fft.rfft(s)
    spsd = (sFFT * sFFT.conj()).real
    spsd = norm**2 * (sFFT * sFFT.conj()).real
    snsd=np.sqrt(spsd)
    
    f1 = open(ofname,"a")
    
    for i in range(1,101):
        Ints=sFFT[Fshake*10*i]
        IntNSDs=snsd[Fshake*10*i]
     
        SigAmpX=IntNSDs
        SigPhaX=np.angle(Ints)         
        f1.write('%f %e %e %e %e %e %e\n' % (Fshake*i, SigAmpX, 0,0, SigPhaX, 0,0))
    f1.close()
```
