```python
def make_psd_self(data_in,fsamp,res):
    data_det=signal.detrend(data_in)
    res = res
    fsamp = fsamp # stays hard coded for now
    freqs=np.linspace(0,int(fsamp/2),(int(res/2))+1)    # change
    FFT=np.fft.rfft(data_det)
    norm = np.sqrt(2 / (res* fsamp))
    psd = norm**2 * (xFFT * FFT.conj()).real
    phase=np.angle(xFFT)
    return freqs,psd,phase

def plot_quick_spectrum(folder,beadnr,data,fsamp,res):
    fname = folder + "/Bead%d" %beadnr + "/InitialTest/Data%d.h5" %data
    BDF = BeadDataFile.BeadDataFile(fname)
    data=BDF.xyz
    trigger = BDF.trigger
    #print(trigger)
    data_det=signal.detrend(data)
    f,xs,xp = make_psd_self(data_in=data_det[0],fsamp=fsamp,res=res)  
    f,ys,yp = make_psd_self(data_in=data_det[1],fsamp=fsamp,res=res)  
    f,zs,zp = make_psd_self(data_in=data_det[2],fsamp=fsamp,res=res)  
    return freqs,xs,ys,zs,xp,yp,zp

def plot_psds(freqs,psd1,psd2,psd3,factor=1000):
    _,ax = plt.subplots()
    ax.loglog(freqs, psd1, '.-', label="x",markersize=5,linewidth=0.5)
    ax.loglog(freqs, psd2, '.-', label="y",markersize=5,linewidth=0.5)
    ax.loglog(freqs, psd3*factor, '.-', label="z",markersize=5,linewidth=0.5)
    ax.set_xlabel('Frequency [Hz]')
    ax.set_ylabel('X PSD [a.u.]')
    #plt.xlim(1,2500)
    #plt.ylim(1e6,1e10)
    plt.grid(True)
    return None
```

```python
folder = "/data/new_trap/20210322/"
beadnr =1
datanr=1
f,x,y,z= plot_quick_spectrum(folder,beadnr,datanr)
#f2,x2,y2,z2= plot_quick_spectrum(folder,beadnr,data)

plot_psds(f,x,y,z*10)#/1e21)
```
