```python
def make_psd_self(data_in,fsamp,res):
    data_det=signal.detrend(data_in)
    res = res
    fsamp = fsamp # stays hard coded for now
    freqs=np.linspace(0,int(fsamp/2),(int(res/2))+1)    # change
    FFT=np.fft.rfft(data_det,n=res)
    norm = np.sqrt(2 / (res* fsamp))
    psd = norm**2 * (FFT * FFT.conj()).real
    phase=np.angle(FFT)
    return freqs,psd,phase

def plot_quick_spectrum(folder,beadnr,data,fsamp,res):
    fname = folder + "/Bead%d" %beadnr + "/InitialTest/Data%d.h5" %data
    BDF = BeadDataFile.BeadDataFile(fname)
    data=BDF.xyz
    trigger = BDF.trigger
    #print(trigger)
    data_det=signal.detrend(data)
    f,xs,xp = make_psd_self(data_in=data_det[0],fsamp=fsamp,res=res)  
    f,ys,yp = make_psd_self(data_in=data_det[1],fsamp=fsamp,res=res)  
    f,zs,zp = make_psd_self(data_in=data_det[2],fsamp=fsamp,res=res)  
    return f,xs,ys,zs,xp,yp,zp

def plot_psds(freqs,psd1,psd2,psd3,factor=100):
    _,ax = plt.subplots()
    ax.loglog(freqs, psd1, '.-', label="x",markersize=1,linewidth=0.5)
    ax.loglog(freqs, psd2*factor*10, '.-', label="y",markersize=1,linewidth=0.5,alpha=1)
    ax.loglog(freqs, psd3*factor, '.-', label="z",markersize=1,linewidth=0.5)
    ax.set_xlabel('Frequency [Hz]')
    ax.set_ylabel('PSD [a.u.]')
    #plt.xlim(1,2500)
    #plt.ylim(1e6,1e10)
    plt.grid(True)
    return None
```

```python
folder = "/data/new_trap/20210324/"
folder1 = "/data/new_trap/20210325/"

beadnr = 4
datanr=2
f,x,y,z,xp,yp,zp= plot_quick_spectrum(folder,beadnr,datanr,5000,5000)
#beadnr2 = 7
#datanr2=0
#f,x2,y2,z2,xp,yp,zp= plot_quick_spectrum(folder,beadnr,datanr,5000,5000)

#f2,x2,y2,z2= plot_quick_spectrum(folder,beadnr,data)

plot_psds(f,x,y,z*10)#/1e21)
#plot_psds(f,x2,y2,z2*10)#/1e21)

plt.ylim(1e1)
print(100+np.argmax(y[100:]))
print(100+np.argmax(x[100:]))
plt.axvline(100+np.argmax(x[100:]))
```
