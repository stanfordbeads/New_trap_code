```python
from skimage.registration import phase_cross_correlation as pcc

def shift_PCC(image,zeroth_image):
    zeroth=np.load(zeroth_image)
    height = 480-np.argmax(np.mean(zeroth,axis=1))
    shift,_,_ = pcc(zeroth,image,upsample_factor=100)
    return shift[0]#+height
```

```python
height1 = []
for image in tqdm(img1):
    height1.append(shift_PCC(image,r"/home/data_transfer/data_from_windows/20201203/Bead2/HeightPictures/HeightMonitor20201203/Pictures/Discharge_0.npy"))
```
