```python
# get file 3

dirPath = "/data/SC_data/ImageData/HeightData/20210423/run3/"
height3FromTxt=np.load(dirPath+"height_info.npy")
time3FromTxt=np.load(dirPath+"time_info.npy")
refPointCorr3 = np.load(dirPath+"image_test_0.npy")
shift3=position_shift_PCC(refPoint0,refPointCorr3)[0]
```

```python
# get file 4

dirPath = "/data/SC_data/ImageData/HeightData/20210423/run/"
height4FromTxt=np.load(dirPath+"height_info.npy")
time4FromTxt=np.load(dirPath+"time_info.npy")
refPointCorr4 = np.load(dirPath+"image_test_0.npy")
shift4=position_shift_PCC(refPoint0,refPointCorr4)[0]
```

```python
totalTime=np.concatenate([time1FromTxt-time1FromTxt[0],
                          time2FromTxt-time1FromTxt[0],
                          time3FromTxt-time1FromTxt[0],
                          time4FromTxt-time1FromTxt[0]])

totalHeight=np.concatenate([height1FromTxt-shift1,height2FromTxt-shift2,height3FromTxt-shift3,height4FromTxt-shift4])
```

```python
dataTimeStart=1619303807612556307/1e9-time1FromTxt[0]
dataTimeEnd=1619355743553861684/1e9-time1FromTxt[0]

bkgTimeStart=1619365382639047096/1e9-time1FromTxt[0]
bkgTimeEnd=1619375730833729179/1e9-time1FromTxt[0]
```

```python
plt.plot(totalTime/3600,totalHeight)
plt.axvspan(dataTimeStart/3600,dataTimeEnd/3600,alpha=0.5,color="orange",label="Shaking")
plt.axvspan(bkgTimeStart/3600,bkgTimeEnd/3600,alpha=0.5,color="purple",label="NoShaking")
plt.legend()
plt.ylabel("pixel shift [~0.5um]")
plt.xlabel("time [h]")
plt.show()
```
