```python
def initial_fft(data,fsamp=5000,res=5000):

    data_det=signal.detrend(data)
    y_1000=data_det[1]
    z_1000=100*data_det[2]

    xpsd, freqs = matplotlib.mlab.psd(data_det[0], Fs = fsamp, NFFT = res)
    ypsd, freqs = matplotlib.mlab.psd(y_1000, Fs = fsamp, NFFT = res)
    zpsd, freqs = matplotlib.mlab.psd(z_1000, Fs = fsamp, NFFT = res)
    print()
    return freqs,xpsd,ypsd,zpsd
```

```python
dataVanilla = BeadDataFile.BeadDataFile(dirName+"vanilla.h5")
dataBitshifted = BeadDataFile.BeadDataFile(dirName+"m3bitshift.h5")
dataNoFb = BeadDataFile.BeadDataFile(dirName+"nofb.h5")
dataFall = BeadDataFile.BeadDataFile(dirName+"afterstabilizing.h5")
dataBR = BeadDataFile.BeadDataFile(dirName+"backtonormal_realigned.h5")
dataAN = BeadDataFile.BeadDataFile(dirName+"backtonormal_realigned_raised.h5")
```
