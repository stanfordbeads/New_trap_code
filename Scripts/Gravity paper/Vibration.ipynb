### Vibration spectrum

```python
bdf_vibration = load_dir(dirname='/data/new_trap/20210503/Bead2/Shaking/NoAttractorShaking0/',
                         file_prefix='NoAttractorShaking0',start_file=4000, max_file=1000)
bdf_vibration2 = bdf_vibration[:1000]

def fft_norm(N, fsamp):
    "Factor to normalize FFT to ASD units"
    return np.sqrt(2 / (N * fsamp))
scale_psd = fft_norm(50000, 5000)

freq = np.fft.rfftfreq(len(bdf_vibration2[0].z2), d=1./5000)
z_psd_vibration = np.abs(np.fft.rfft(bdf_vibration2[0].z2))
for bb in bdf_vibration2[1:]:
    z_psd_vibration += np.abs(np.fft.rfft(bb.z2))
z_psd_vibration /= len(bdf_vibration2)
z_psd_vibration *= scale_psd
z_psd_vibration /= gfw.scale_Z2

freq = np.fft.rfftfreq(len(bdf_vibration2[0].z2), d=1./5000)
x_psd_vibration = np.abs(np.fft.rfft(bdf_vibration2[0].x2))
for bb in bdf_vibration2[1:]:
    x_psd_vibration += np.abs(np.fft.rfft(bb.x2))
x_psd_vibration /= len(bdf_vibration2)
x_psd_vibration *= scale_psd
x_psd_vibration /= gfw.scale_X2
```

```python
_,ax = plt.subplots(figsize=(9,4))
ax.semilogy(freq, x_psd_vibration, label='X - NoAttractorShaking6 - 20210423')
ax.legend()
_ = ax.set(xlim=(0,50), ylim=(1e-22,2e-15), xlabel='Frequency [Hz.]', ylabel='Force [N]', title=r'X force')
ax.grid()
```

```python
_,ax = plt.subplots(figsize=(9,4))
ax.semilogy(freq, z_psd_vibration, label='Z - NoAttractorShaking6 - 20210423')
ax.legend()
_ = ax.set(xlim=(0,50), ylim=(5e-18,2e-15), xlabel='Frequency [Hz.]', ylabel='Force [N]', title=r'Z force')
ax.grid()
```
