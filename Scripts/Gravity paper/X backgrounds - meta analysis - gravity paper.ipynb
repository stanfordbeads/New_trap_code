```python
def get_xt_freq(freq_):
    res1 = []
    res2 = []
    for i in range(1,100):
#         fname = r'/data/new_trap/20200107/Bead3/Shaking/Shaking360/Shaking6_'+str(i)+'.h5'
#         fname = r'/data/new_trap/20200107/Bead3/Shaking/Shaking340/Shaking8_'+str(i)+'.h5'
        fname = r'/data/new_trap/20200107/Bead3/Shaking/Shaking380/Shaking4_'+str(i)+'.h5'
#     fname = r'/data/new_trap/20200107/Bead3/Shaking/Shaking350/Shaking7_'+str(i)+'.h5'
#         fname = r'/data/new_trap/20200107/Bead3/Shaking/Shaking370/Shaking5_'+str(i)+'.h5'

        bb = BeadDataFile.BeadDataFile(fname=fname)
        res1.append((np.std(bb.response_at_freq2('x', drive_freq=freq_))*50000/gfw.scale_X2))
        res2.append(-1*(np.std(bb.response_at_freq2('z', drive_freq=freq_))/gfw.scale_Z2))
    return res1, res2
```
