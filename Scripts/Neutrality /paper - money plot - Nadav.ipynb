```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
neutrality_iron, mass_iron*6e23
```

```python
mass_iron = 2e-4
charge_iron = 1e-1
neutrality_iron = charge_iron/(mass_iron*6e23)

charge_us = 3.5e-5
mass_us = 420e-12
neutrality_us = charge_us/(mass_us*6e23)

charge_us2 = 3.5e-5/10/5
mass_us2 = 420e-12*(20/7.5)**3
neutrality_us2 = charge_us2/(mass_us2*6e23)

charge_dave = 1e-4
mass_dave = mass_us/(1.5)**3
neutrality_dave = charge_dave/(mass_dave*6e23)


charge_gadi = np.array([2e-4, 6e-4, 11e-4])
mass_gadi = mass_us*np.array([(10/7.5)**3, (15/7.5)**3, (20/7.5)**3])
neutrality_gadi = charge_gadi/(mass_gadi*6e23)
```

```python
range_plot = np.logspace(-7,0)
neutality_21 = range_plot/1e-21/6e23
neutality_19 = range_plot/1e-19/6e23
neutality_23 = range_plot/1e-23/6e23
```

```python
_,ax = plt.subplots(figsize=(8,6))
ax.loglog(range_plot, neutality_19, '--', label=r'$10^{-19}$')
ax.loglog(range_plot, neutality_21, '--', label=r'$10^{-21}$')
ax.loglog(range_plot, neutality_23, '--', label=r'$10^{-23}$')
ax.set(xlabel='Charge sensitivity [e]', ylabel='Mass [g]',xlim=(1e-7,0.5), ylim=(1e-11,0.1))
ax.scatter(charge_iron, mass_iron, s=100, marker='*')
ax.scatter(charge_us, mass_us, s=100, marker='o')
ax.scatter(charge_us2, mass_us2, s=100, marker='o')
ax.scatter(charge_dave, mass_dave, s=100, marker='v')
ax.scatter(charge_gadi, mass_gadi, s=100, marker='p')
ax.legend()
ax.invert_xaxis()
```

```python
_,ax = plt.subplots(figsize=(8,6))
ax.loglog(range_plot, neutality_19*0+1e-19, '--')
ax.loglog(range_plot, neutality_19*0+1e-20, '--')
ax.loglog(range_plot, neutality_21*0+1e-21, '--')
ax.loglog(range_plot, neutality_23*0+1e-22, '--')
ax.set(xlabel='Charge sensitivity [e]', ylabel='$|q_{p}+q_{e}|$  [e]',xlim=(1e-7,0.5))
ax.scatter(charge_iron, neutrality_iron,  s=100, marker='*', label='iron spheres')
ax.scatter(charge_us, neutrality_us, s=100, marker='o', label='us')
ax.scatter(charge_us2, neutrality_us2,  s=100, marker='o', label='future')
ax.scatter(charge_dave, neutrality_dave,  s=100, marker='v', label='Stanford 2014')
ax.scatter(charge_gadi, neutrality_gadi,  s=100, marker='p', label='Yale')
ax.legend()
ax.invert_xaxis()
```
