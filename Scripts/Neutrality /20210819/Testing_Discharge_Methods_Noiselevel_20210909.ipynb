```python
def digital_demod(dat, fDemod, fs=5000, tFFT=10, win=('tukey',0.25),
        nOverlap=0, detrend='constant', median=False):
    '''
    Single frequency digital demodulation.

    Parameters:
    -----------
    dat: array_like
         Data to demodulate. May contain multiple time series
    fDemod: float
         Frequency at which to do the demodulation.
    fs: float, optional
        Sampling frequency of the time series. Defaults to 2**14=16384 Hz.
    tFFT: float, optional
        Segment length (seconds) to evaluate the FFT. Defaults to 5 s.
    win: tuple, optional
        Input to scipy.signal window function. Defaults to Tukey window with alpha=0.25
    nOverlap: int, optional
        Number of samples to overlap window. Defaults to 0.
    detrend: string, optional
        Input to scipy.signal detrend function. Defaults to 'constant'
    median: Bool, optional
        Median averaging of final result. Defaults to False.

    Returns:
    --------
    result: complex
        Result of the digital demodulation.
    TODO: error handling...
'''
    band_width=2
    b, a = signal.butter(3, [2.*(fdrive-band_width/2.)/fsamp, \
                         2.*(fdrive+band_width/2.)/fsamp ], btype = 'bandpass')
    dat = signal.filtfilt(b, a, dat)
    dat = np.asarray(dat)
    if dat.size==0:
        return(np.empty(dat.shape[-1]))
    nperseg = int(np.round(tFFT*fs)) # Number of segments in a sample segment.
    nOverlap = int(nOverlap);
    # Make the LO time series
    tt = np.arange(len(dat))/fs
    LO = np.exp(-1j*2*np.pi*fDemod*tt)

    # Compute the step to take as we stride through the segments
    step = nperseg - nOverlap
    segShape = ((dat.shape[-1]-nOverlap)//step, nperseg)
    datStrides = (step*dat.strides[-1], dat.strides[-1])
    LOStrides = (step*LO.strides[-1], LO.strides[-1])
    dS = np.lib.stride_tricks.as_strided(dat, shape=segShape, strides=datStrides)
    LOS = np.lib.stride_tricks.as_strided(LO, shape=segShape, strides=LOStrides)

    # Detrend the data
    data = signal.detrend(dS, type=detrend)

    # Demodulate the (windowed) data
    wind = signal.get_window(win,nperseg)
    result = data * wind * LOS
    result = result.sum(axis=-1)
    scale = 2*np.sqrt(1/(wind.sum()**2))
    result *= scale
    return result
```

```python
def check_method_digitalDemod(noiseLevel,iterations,signalIn=sigIn):
    corrList = []
    corrSideList = []
    for i in range(iterations):
        whiteNoise=np.random.normal(0,noiseLevel,nsamp)
        totSignal=whiteNoise+signalIn
        demod = digital_demod(dat=totSignal,fDemod=71,fs=5000,tFFT=10)
        demodSide = digital_demod(dat=totSignal,fDemod=73,fs=5000,tFFT=10)
        corrList.append(demod.imag)
        corrSideList.append(demodSide.imag)
    return [np.mean(corrList),np.std(corrList)],[np.mean(corrSideList),np.std(corrSideList)]
```

```python
def check_method_corr(noiseLevel,iterations,signalIn=sigIn):
    corrList = []
    corrSideList = []
    for i in range(iterations):
        whiteNoise=np.random.normal(0,noiseLevel,nsamp)
        totSignal=whiteNoise+signalIn
        corrs = correlation(drive=signalIn, response=totSignal, fsamp=fsamp, fdrive=fdrive, filt=True, band_width=2)
        corrsSide = correlation(drive=signalIn, response=totSignal, fsamp=fsamp, fdrive=fdrive+2, filt=True, band_width=2)
        corrList.append(corrs[0])
        corrSideList.append(corrsSide[0])
    return [np.mean(corrList),np.std(corrList)],[np.mean(corrSideList),np.std(corrSideList)]
```

```python
def check_method_sine(noiseLevel,iterations,signalIn=sigIn):
    drive_freq=71
    bandwidth=2
    decimate=10
    #calibFreqs = np.fft.rfftfreq(len(sig), d=1./5000) # fft
    #fft_angles = np.angle(np.fft.rfft(sig)) # get the phase
    #phaseResponse = (fft_angles[calibFreqs==drive_freq])%(2*np.pi)
    corrList=[]
    corrSideList = []
    
    for i in tqdm(range(iterations)):
        ll = likelihood_analyser.LikelihoodAnalyser()
        whiteNoise=np.random.normal(0,noiseLevel,nsamp)
        totSignal=whiteNoise+signalIn

        fit_kwargs = {'A': 0, 'f': drive_freq, 'phi': 0, 'sigma': 0.1,
                  'error_A': 1, 'error_f': 1, 'error_phi': 0.5, 'errordef': 1, 'error_sigma':0.1,
                  'limit_phi': [-2 * np.pi, 2 * np.pi], 'limit_sigma': [0, 1],
                  'limit_A': [-100, 100], 
                  'print_level': 0, 'fix_f': True, 'fix_phi': True}


        b, a = signal.butter(3, [2.*(drive_freq-bandwidth/2.)/fsamp, 2.*(drive_freq+bandwidth/2.)/fsamp ], btype = 'bandpass')
        res = signal.filtfilt(b, a, totSignal)[::decimate]

        m1_tmp = ll.find_mle_sin(res, fsamp=5000/decimate, noise_rms=1, plot=False, suppress_print=True, **fit_kwargs)
        
        b2, a2 = signal.butter(3, [2.*(drive_freq+2-bandwidth/2.)/fsamp, 2.*(drive_freq+2+bandwidth/2.)/fsamp ], btype = 'bandpass')
        res2 = signal.filtfilt(b, a, totSignal)[::decimate]

        m1_tmp2 = ll.find_mle_sin(res2, fsamp=5000/decimate, noise_rms=1, plot=False, suppress_print=True, **fit_kwargs)

        corrList.append(m1_tmp.values[0])
        corrSideList.append(m1_tmp2.values[0])
    return [np.mean(corrList),np.std(corrList)],[np.mean(corrSideList),np.std(corrSideList)]
```
