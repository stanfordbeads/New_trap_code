## Force calibration - X

```python
foldername = r'/home/data_transfer/data_from_windows/20190701/Bead2/Discharge2/'
BDFs = load_dir(foldername, file_prefix='Discharge')
```

```python
freq = 23
responsesX = [B.response_at_freq('x',drive_freq=freq) for B in BDFs]
rrX = [np.std(resp_) for resp_ in responsesX]
_,ax = plt.subplots(figsize=(9,4))
ax.scatter(range(len(rrX)),rrX)
ii = [4,5]
ij = [0,1,6,7]
SNR = np.mean([rrX[i] for i in ii])/np.mean([rrX[i] for i in ij])
print(1.6e-19*20/8.6e-3/SNR*0.65, SNR)
```
