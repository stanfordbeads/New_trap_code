```python
import sys
sys.path.append('../../nadavp/new_trap/PythonScripts/')
import BeadDataFile
from discharge_tools import *
```

```python
fname_4 = r'/data/new_trap/20190708/Bead2/Discharge/Discharge_729.h5'
bd_4 = BeadDataFile.BeadDataFile(fname_4)
freq,psd = bd_4.psd2('y')
peak_response = bd_4.response_at_freq2('y', drive_freq = 23,bandwidth = 1)
```

```python
_,ax = plt.subplots(1,3,figsize=(15,5))
ax[0].loglog(freq,psd)
ax[0].set(xlim=(0,3000))
ax[0].set_title('PSD over 10 sec')
ax[1].set_title('Time Domain - All Frequencies')
ax[2].plot(range(50000),peak_response)
ax[1].plot(range(50000),bd_4.y2)
ax[2].set_title('Time Domain - 1Hz Bandpass around 23Hz')
```

```python
driven_data = bd_4.y2[:25000]
relaxation_data = bd_4.y2[25000:25500]
driven_psd = np.abs(np.fft.rfft(driven_data))**2
relaxation_psd = np.abs(np.fft.rfft(relaxation_data))**2
driven_freq = np.fft.rfftfreq(len(driven_data),d = 1/bd_4.fsamp)
relaxation_freq = np.fft.rfftfreq(len(relaxation_data), d = 1/bd_4.fsamp)
```

```python
_,ax = plt.subplots(1,2,figsize = (10,5))
ax[0].loglog(driven_freq, driven_psd)
ax[1].loglog(relaxation_freq, relaxation_psd)
#ax[1].set(xlim = (75,250))
#ax[0].set(xlim = (75,250))
```

# Short Time Fourier Transform

```python
f, t, Zxx = signal.stft(bd_4.y2, fs = 5000, nperseg = 500,noverlap = None)
```

```python
fig,ax = plt.subplots(figsize= (9,5))
ax.pcolormesh(t,f,np.abs(Zxx))
ax.set(ylim = (0,600),ylabel = 'Frequency (Hz)', xlabel = 'Time (s)')
fig.colorbar(ax.pcolormesh(t,f,np.abs(Zxx)),label = 'Magnitude of PSD')
```
