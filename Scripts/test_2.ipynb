### Ringdown

```python
A = 100.
f = 100
f1 = 150
t = np.arange(0,10,0.0002)
samples = 10/0.0002
# ff = np.arange(f,f1,(f1-f)/samples)
ff = np.logspace(np.log(f),np.log(f1),num=50000, base=np.exp(1))
# ff = f1
data = A*np.sin(2*np.pi*t*ff)
```

```python
ypsd, freqs = mlab.psd(data, Fs = 5000, NFFT=50000)
x = data
fft2 = np.abs(np.fft.rfft(x))**2
freq2 = np.fft.rfftfreq(len(x), d=1./5000)
```

```python
_,ax = plt.subplots()
ax.semilogy(freqs, ypsd/ypsd[freqs==1000])
ax.semilogy(freq2, fft2/fft2[freq2==1000])
# ax.set(xlim=(f-10,f1+10),ylim=(1e-4,None))
```
