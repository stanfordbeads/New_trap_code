### Load and process multpule 1000 datasets

```python
fit_kwargs2 = {'alpha': 0, 'phase': 0, 'sigma': 17, 'error_alpha': 0.001, 'error_phase': 1, 'error_sigma': 1,
              'errordef': 1,
              'limit_alpha': [-1, 1], 'limit_phase': [-3,3], 'limit_sigma': [0,None],
              'fix_alpha': False, 'fix_phase': False, 'fix_sigma': False, 'print_level': 0}


fit_kwargs2_x = {'alpha': 0, 'phase': 0, 'sigma': 17, 'error_alpha': 0.001, 'error_phase': 1, 'error_sigma': 1,
              'errordef': 1,
              'limit_alpha': [-20, 20], 'limit_phase': [-3,3], 'limit_sigma': [0,None],
              'fix_alpha': False, 'fix_phase': False, 'fix_sigma': False, 'print_level': 0}

min_18_z = []
min_18_x = []
min_10_z = []
datasets_num = 1000
bandwidth = 2
center_freq = 18
noise_freq = 22.5
# Normal feedback values
foldername = r'/data/new_trap/20200226/Bead1/Shaking/Shaking385/'
datasets_num = 1000
Nj = 23

for j in range(Nj):
    
    t = time.time()
    print(j,'/',Nj)
    del gfw.BDFs
    gfw.BDFs = load_dir(foldername, file_prefix='Shaking3', start_file=j*1000, max_file=datasets_num)
    
    for i,bb in enumerate(gfw.BDFs):
        try:
            with io.capture_output() as captured:
                min_18_z.append(gfw.get_alpha_mle_pl(bdf=bb, center_freq=center_freq, noise_freq=noise_freq, 
                                                     bandwidth=bandwidth, decimate=100, direction1='z',
                                  x_focous=393.17, frequency=3, offset_y=10e-6, lambda_par=0.0001, height=6e-6, suppress_print=False,
                                  **fit_kwargs2)[2])
    #             min_10_z.append(gfw.get_alpha_mle_pl(bdf=bb, center_freq=20, noise_freq=10, 
    #                                              bandwidth=bandwidth, decimate=100, direction1='z',
    #                           x_focous=393.17, frequency=5, lambda_par=0.0001, height=6e-6, suppress_print=False,
    #                           **fit_kwargs2)[2])
                min_18_x.append(gfw.get_alpha_mle_pl(bdf=bb, center_freq=center_freq, noise_freq=noise_freq, 
                                                     bandwidth=bandwidth, decimate=100, direction1='x',
                                  x_focous=393.17, frequency=3, offset_y=10e-6, lambda_par=0.0001, height=6e-6, suppress_print=False,
                                  **fit_kwargs2_x)[2])
        except:
            min_18_z.append(0)
            min_18_x.append(0)
    print('elapsed time: ',time.time() - t)
```
