```python
fit_kwargs = {'alpha': 1e-2, 'phase': 0, 'error_alpha': 0.01, 'error_phase': 1,
              'errordef': 1,
              'limit_alpha': [1e-5, 10], 'limit_phase': [-5000,5000],
              'fix_alpha': False, 'fix_phase': False, 'print_level': 0}
min_18_xz = []
datasets_num = 350
for i,bb in enumerate(gfw.BDFs):
    print(i,'/',datasets_num)
    min_18_xz.append(gfw.get_alpha_2d(bdf=bb, center_freq=18, bandwidth=2, direction1='x', direction2='z',
                          x_focous=389.0, frequency=3, lambda_par=0.0001, height=6e-6, decimate=10, suppress_print=False,
                          **fit_kwargs)[2])
```

```python
m_18_z = [m_.values[0] for m_ in min_18_z]
m_18_x = [m_.values[0] for m_ in min_18_x]
m_18_xz = [m_.values[0] for m_ in min_18_xz]
m_18_noise = [m_.values[0] for m_ in min_18_noise]

_, ax = plt.subplots(figsize=(9,4))
ax.semilogy(np.arange(350), np.array(m_18_noise)*1e10, '.', label=r'Z noise')
ax.semilogy(np.arange(350), np.array(m_18_x)*1e10, '.', label=r'X only')
ax.semilogy(np.arange(350), np.array(m_18_z)*1e10, '.', label=r'Z only')
ax.semilogy(np.arange(350), np.array(m_18_xz)*1e10, '.', label=r'X & Z')
ax.legend()
print('mean alpha (Z-only): ', np.mean(m_18_z), 'std: ', np.std(m_18_z)/np.sqrt(350)*2)
print('mean alpha (Z-only): ', np.mean(m_18_x), 'std: ', np.std(m_18_x)/np.sqrt(350)*2)
print('mean alpha (Z-only) Noise: ', np.mean(m_18_noise), 'std: ', np.std(m_18_noise)/np.sqrt(350)*2)
```

```python
_, ax = plt.subplots(1,2,figsize=(9.5,4))
ax[0].semilogy(np.arange(350), np.array(m_18_z)*1e10, '.', label=r'6th harmonic')
ax[0].semilogy(np.arange(350), np.array(m_18_noise)*1e10, '.', label=r'noise')
ax[0].set(ylim=(1e7,5e10), xlabel='time [sec.]', ylabel=r'$\alpha$', title=r'$\lambda=100 \mu m $')
ax[0].legend()
ax[1].hist(m_18_z,bins=20,alpha=0.5)
ax[1].hist(m_18_noise,bins=20,alpha=0.5)
```
