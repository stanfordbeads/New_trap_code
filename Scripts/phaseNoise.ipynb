```python
from nptdms import TdmsFile
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
import scipy.signal as sig
```

```python
fftDict = {}
tFFT = 10 # seconds, for averaging
fftDict['fs'] = 10e3
fftDict['nperseg'] = int(tFFT*10e3)
fftDict['window'] = sig.get_window(('tukey',0.25),fftDict['nperseg'])

V2rad = np.pi/15e-3 # Convert V/rtHz to rad/rtHz
```

```python
ADCnoise = TdmsFile('/home/data_transfer/data_from_windows/Misc/SRSphaseNoise/ADCnoise.tdms')
mixerNoise = TdmsFile('/home/data_transfer/data_from_windows/Misc/SRSphaseNoise/mixerNoise.tdms')
phaseNoise = TdmsFile('/home/data_transfer/data_from_windows/Misc/SRSphaseNoise/phaseNoise.tdms')
phaseNoise_ref = TdmsFile('/home/data_transfer/data_from_windows/Misc/SRSphaseNoise/phaseNoise_ref.tdms')
# drivenNoise_ref = TdmsFile('/home/data_transfer/data_from_windows/Misc/SRSphaseNoise/drivenNoise_ref.tdms') # 1kHz band, 400kHz RMS
drivenNoise_ref = TdmsFile('/home/data_transfer/data_from_windows/Misc/SRSphaseNoise/drivenNoise2_ref.tdms') # 1kHz band, 70deg RMS


xx_ADC = ADCnoise['ai']['phaseNoise0']
xx_mixer = mixerNoise['ai']['phaseNoise0']
xx_phaseNoise = phaseNoise['ai']['phaseNoise0']
xx_phaseNoiseRef = phaseNoise_ref['ai']['phaseNoise0']
xx_drivenNoiseRef = drivenNoise_ref['ai']['phaseNoise0']


ff_ADC, Pxx_ADC = sig.welch(xx_ADC, **fftDict)
ff_mixer, Pxx_mixer = sig.welch(xx_mixer, **fftDict)
ff_phaseNoise, Pxx_phaseNoise = sig.welch(xx_phaseNoise, **fftDict)
ff_phaseNoiseRef, Pxx_phaseNoiseRef = sig.welch(xx_phaseNoiseRef, **fftDict)
ff_drivenNoiseRef, Pxx_drivenNoiseRef = sig.welch(xx_drivenNoiseRef, **fftDict)
```

```python
fig, ax = plt.subplots(1,1, figsize=(16,9))
ax.loglog(ff_ADC, np.sqrt(Pxx_ADC)*V2rad, rasterized=True, alpha=0.7, label='ADC noise') 
ax.loglog(ff_mixer, np.sqrt(Pxx_mixer)*V2rad, rasterized=True, alpha=0.7, label='Mixer (RF terminated)') 
ax.loglog(ff_phaseNoise, np.sqrt(Pxx_phaseNoise)*V2rad, rasterized=True, alpha=0.7, label='Phase noise (Trapping Beam)')
ax.loglog(ff_phaseNoiseRef, np.sqrt(Pxx_phaseNoiseRef)*V2rad, rasterized=True, alpha=0.7, label='Phase noise (Ref Beam)')
ax.loglog(ff_drivenNoiseRef, np.sqrt(Pxx_drivenNoiseRef)*V2rad, rasterized=True, alpha=0.7, label='Driven noise (Ref Beam)')

ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Phase noise')
ax.legend(loc='best')
```
