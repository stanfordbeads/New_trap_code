```python
def reduced_df(df_in,parameter=["amplitude_x","amplitude_y","amplitude_z","phase_x","phase_y","phase_z"]):
    df_red = pd.DataFrame()
    for elements in tqdm(parameter):
        for j in range(15):
            list_temp = []
            for i in range(len(df_in)):
                list_temp.append(df_in["%s" %elements][i][j])
            df_red["%s_%d" %(elements,j*3+3)] = list_temp
    df_red=df_red.reset_index()
    return df_red
```

```python
def compare_xzyspin_with_pearson(corr_par,corr_val,pars,df_data,df_spin,corr_with,spin=True):
    par_of_interest = corr_val
    x_list = []
    y_list = []
    spin_list = []
    z_list = []
    if(spin==True): df = df_spin
    if(spin==False): df =df_data    

    for i in pars:
        x_list.append(df["%s_%d" %(corr_par,par_of_interest)].corr(df_data["%s_%d" %(corr_with[0],i)]))
        y_list.append(df["%s_%d" %(corr_par,par_of_interest)].corr(df_data["%s_%d" %(corr_with[1],i)]))
        z_list.append(df["%s_%d" %(corr_par,par_of_interest)].corr(df_data["%s_%d" %(corr_with[2],i)]))
        spin_list.append(df["%s_%d" %(corr_par,par_of_interest)].corr(df_spin["%s_%d" %(corr_with[3],i)]))
    x = np.arange(3,48,3)
    print("Freq:",par_of_interest)
    print(np.argmax(x_list),(np.argmin(x_list)))
    print(np.argmax(y_list),(np.argmin(y_list)))
    print(np.argmax(z_list),(np.argmin(z_list)))
    print(np.argmax(spin_list),(np.argmin(spin_list)))
    plt.title("background correlation with %s_%d" %(corr_par,par_of_interest))
    plt.plot(x,x_list,marker="$X$",linestyle="--",markersize=12,label="%s" %corr_with[0])
    plt.plot(x,z_list,marker="$Z$",linestyle="--",markersize=12,label="%s" %corr_with[1])
    plt.plot(x,y_list,marker="$Y$",linestyle="--",markersize=12,label="%s" %corr_with[2])
    plt.plot(x,spin_list,marker="$S$",linestyle="--",markersize=12,label="%s" %corr_with[3])
    plt.axhline(0, color ="red")
    plt.ylabel("pearson coefficient")
    plt.xlabel("frequency")
    plt.legend()
    plt.show()
    return None
```

```python
corr_with = ["amplitude_x","amplitude_y","amplitude_z","freq"]
compare_xzyspin_with_pearson(corr_par=str(corr_with[0]),corr_val=9,pars=np.arange(3,48,3),df_data=df_data_red,df_spin=df_spin_red,corr_with = corr_with,spin=False)
```
